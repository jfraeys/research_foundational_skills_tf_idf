```python
df_soft_skills = pd.read_csv(
    "data/external/taxosoft_soft_skills/taxosoft_soft_skills.csv", header=0
)
```

```python
df_soft_skills.rename(
    columns={"Soft skill": "soft_skill", "Alternative labels": "alt_labels"},
    inplace=True,
)
```

```python
df_soft_skills = df_soft_skills.convert_dtypes()
```

```python
df_soft_skills["soft_skill"] = df_soft_skills["soft_skill"].str.lower()
df_soft_skills["alt_labels"] = df_soft_skills["alt_labels"].str.lower()
```

```python
df_soft_skills.sample(3).to_csv('reports/sample3_found_skills.csv', index=False)
```
